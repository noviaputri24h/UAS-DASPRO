```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Students Social Media Addiction.csv")

# Kelompokkan berdasarkan jenjang akademik dan hitung rata-rata
avg_df = df.groupby('Academic_Level')[[
    'Avg_Daily_Usage_Hours',
    'Sleep_Hours_Per_Night',
    'Mental_Health_Score',
    'Addicted_Score'
]].mean().reset_index()

# Buat line chart
plt.figure(figsize=(12, 8))

for column in ['Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 'Mental_Health_Score', 'Addicted_Score']:
    plt.plot(avg_df['Academic_Level'], avg_df[column], marker='o', label=column)

plt.title('Rata-rata Statistik Sosial Mahasiswa berdasarkan Jenjang Akademik')
plt.xlabel('Jenjang Akademik')
plt.ylabel('Skor / Jam')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=-45)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Students Social Media Addiction.csv")

# Hitung frekuensi platform
platform_counts = df['Most_Used_Platform'].value_counts()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(platform_counts, labels=platform_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Platform Media Sosial yang Paling Sering Digunakan')
plt.axis('equal')  # Biar lingkarannya bulat
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Students Social Media Addiction.csv")

avg_usage = df.groupby('Country')['Avg_Daily_Usage_Hours'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
avg_usage.plot(kind='bar', color='skyblue')
plt.title('Rata-rata Jam Penggunaan Media Sosial per Hari (Top 10 Negara)')
plt.xlabel('Negara')
plt.ylabel('Jam / Hari')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
